```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load cleaned data
print("Loading cleaned dataset...")
df = pd.read_csv("/Users/faribakazi/Documents/Cyclistic_Bike_Share_Analysis/data/processed/cyclistic_cleaned.csv")

print(f"Total rides: {len(df):,}")
print(f"Date range: {df['started_at'].min()} to {df['started_at'].max()}")

# KEY INSIGHTS
print("\n=== KEY INSIGHTS ===")
print("Member vs Casual distribution:")
print(df['member_casual'].value_counts(normalize=True).round(4))

print("\nAverage ride length (minutes):")
print(df.groupby('member_casual')['ride_length'].mean().round(2))

# VISUALIZATIONS
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)

# 1. Average Ride Length by Rider Type
plt.figure()
sns.barplot(x='member_casual', y='ride_length', data=df, ci=None, palette=['#1f77b4', '#ff7f0e'])
plt.title('Average Ride Length by Rider Type')
plt.xlabel('Rider Type')
plt.ylabel('Average Ride Length (minutes)')
plt.savefig("/Users/faribakazi/Downloads/avg_ride_length.png", dpi=300, bbox_inches='tight')
plt.show()

# 2. Number of Rides by Day of Week
plt.figure()
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
df['day_name'] = pd.Categorical(df['day_name'], categories=day_order, ordered=True)
sns.countplot(x='day_name', hue='member_casual', data=df, palette=['#1f77b4', '#ff7f0e'])
plt.title('Number of Rides by Day of Week')
plt.xticks(rotation=45)
plt.legend(title='Rider Type')
plt.savefig("/Users/faribakazi/Downloads/rides_by_day.png", dpi=300, bbox_inches='tight')
plt.show()

# 3. Average Ride Length by Day of Week
plt.figure()
sns.barplot(x='day_name', y='ride_length', hue='member_casual', data=df, ci=None, palette=['#1f77b4', '#ff7f0e'])
plt.title('Average Ride Length by Day of Week')
plt.xticks(rotation=45)
plt.legend(title='Rider Type')
plt.savefig("/Users/faribakazi/Downloads/ride_length_by_day.png", dpi=300, bbox_inches='tight')
plt.show()

# 4. Rides by Hour of Day
plt.figure()
sns.histplot(data=df, x='hour_of_day', hue='member_casual', bins=24, multiple='dodge', palette=['#1f77b4', '#ff7f0e'])
plt.title('Rides by Hour of Day')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Rides')
plt.legend(title='Rider Type')
plt.savefig("/Users/faribakazi/Downloads/rides_by_hour.png", dpi=300, bbox_inches='tight')
plt.show()

print("\n All visualizations saved to Downloads folder.")
print("Python notebook is complete.")
```
